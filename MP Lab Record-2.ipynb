```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io
import skimage.color as cl

img = io.imread("naruto.png",0)
img_gr = io.imread("naruto.png",1)
img_fft = np.fft.fftshift(np.fft.fft2(img_gr))
img_recons = abs(np.fft.ifft2(img_fft))
t = ["Original Image","Grayscaled Image","Magnituded Spectrum","Reconstructed Image"]
im = [img,img_gr,np.log(abs(img_fft)),img_recons]
for i in range(1):
    plt.subplot(2,2,i+1)
    plt.title(t[i])
    plt.imshow(im[i])
    plt.axis('off')
for i in range(1,4):
    plt.subplot(2,2,i+1)
    plt.title(t[i])
    plt.imshow(im[i],cmap="gray")
    plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2 as cv

files='.jpg'
img=[]
for i in range(4):
    n=str(i+1)
    img.append(n+files)
th = [70,140,125,127,120]

for i in range(4):
  img1 = cv.imread(img[i], 0)
  out, t1 = cv.threshold(img1, th[i], 255, cv.THRESH_BINARY)
  out, t2 = cv.threshold(img1, th[i], 255, cv.THRESH_BINARY_INV)
  out, t3 = cv.threshold(img1, th[i], 255, cv.THRESH_TRUNC)
  out, t4 = cv.threshold(img1, th[i], 255, cv.THRESH_TOZERO)
  out, t5 = cv.threshold(img1, th[i], 255, cv.THRESH_TOZERO_INV)
  fig, ax = plt.subplots(1, 6, figsize=(20, 10))
  plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
  ax[0].imshow(img1, cmap="gray")
  ax[0].set_title("Original Image")
  ax[1].imshow(t1, cmap="gray")
  ax[1].set_title("Binary Image")
  ax[2].imshow(t2, cmap="gray")
  ax[2].set_title("Inverse of Binary Image")
  ax[3].imshow(t3, cmap="gray")
  ax[3].set_title("Truncated Image")
  ax[4].imshow(t4, cmap="gray")
  ax[4].set_title("0 Image")
  ax[5].imshow(t5, cmap="gray")
  ax[5].set_title("Inverse of 0 Image")
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2 as cv

files='.jpg'
img=[]
for i in range(4):
    n=str(i+1)
    img.append(n+files)
th = [70,140,125,127,120]

for i in range(4):
  img1 = cv.imread(img[i], 0)
  t1 = cv.adaptiveThreshold(img1,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,9,2)
  t2 = cv.adaptiveThreshold(img1,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,9,2)
  out, t3 = cv.threshold(img1,80,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU)
  fig, ax = plt.subplots(1, 3, figsize=(15, 7.5))
  ax[0].imshow(t1, cmap="gray")
  ax[0].set_title("Adaptive thershold")
  ax[1].imshow(t2, cmap="gray")
  ax[1].set_title("Adaptive thershold gaussian")
  ax[2].imshow(t3, cmap="gray")
  ax[2].set_title("Adaptive thershold OTSU")
  plt.show()
```
