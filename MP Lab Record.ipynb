```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imshow, imread
from skimage.transform import downscale_local_mean

img = imread('naruto.png')
factors = 3**np.arange(1, 5)
figure,axis= plt.subplots(1, len(factors), figsize=(15,3))

for factor, ax in zip(factors, axis):
    image = downscale_local_mean(img,factors=(factor, factor, 1)).astype(int)
    ax.imshow(image)
    ax.set_title('$N={}$'.format(image.shape[0]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

def square_image(x, y):
  X, Y = np.meshgrid(x, y)
  return X**2 + Y**2

factors = 2**np.arange(1, 5)
squ_image = square_image(np.linspace(0, 1, num=5), np.linspace(0, 
1, num=5))
fig, ax = plt.subplots(1, len(factors)+1, figsize=(15, 4))
ax[0].imshow(squ_image)
ax[0].set_title("Original Image")

for i, k in enumerate(factors):
  i=i+1
  bins = np.linspace(0, squ_image.max(), k)
  image = np.digitize(squ_image, bins)
  image = np.vectorize(bins.tolist().__getitem__)(image-1)
  ax[i].imshow(image)
  ax[i].set_title('$k = {}$'.format(k))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread("naruto.png",flags=2)
hist1,bins = np.histogram(img.flatten(),256,[0,256])
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
equ = cv.equalizeHist(img)
hist2,bins = np.histogram(equ.flatten(),256,[0,256])
plt.hist(equ.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])

t = ["Original Image","Equalized Image","Original Image [Histogram]","Equalized Image [Histogram]"]
im = [img,equ,hist1,hist2]
for i in range(2):
    plt.subplot(2,2,i+1)
    plt.title(t[i])
    plt.imshow(im[i],cmap='gray')
    plt.axis('off')
for i in range(2,3):
    plt.subplot(2,2,i+1)
    plt.title(t[i])
    plt.plot(bin[0:256],hist1)
for i in range(3,4):
    plt.subplot(2,2,i+1)
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
    plt.title(t[i])
    plt.plot(bin[0:256],hist2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io
img= io.imread("naruto.png")
img1=cv.medianBlur(img,7)
img2 = cv.blur(img,(7,7),borderType=cv.BORDER_CONSTANT)
kernel=np.ones((5,5),np.float32)/25
img3=cv.filter2D(img,-1,kernel)
img4=cv.GaussianBlur(img,(7,7),0.5)
img5=cv.bilateralFilter(img,9,50,10)
t=["Orginal Image","Medium Blur","Blur Image","Filter 2D","Gaussian Blur","Bilateral Filter"]
im=[img,img1,img2,img3,img4,img5]
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.title(t[i])
    plt.imshow(im[i])
    plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import skimage.io as io
import skimage.color as cl
img = io.imread("naruto.png",0)
img_gr = io.imread("naruto.png",1)
img_fft = np.fft.fftshift(np.fft.fft2(img_gr))
img_recons = abs(np.fft.ifft2(img_fft))
t = ["Original Image","Grayscaled Image","Magnituded Spectrum","Reconstructed Image"]
im = [img,img_gr,np.log(abs(img_fft)),img_recons]
for i in range(1):
    plt.subplot(2,2,i+1)
    plt.title(t[i])
    plt.imshow(im[i])
    plt.axis('off')
for i in range(1,4):
    plt.subplot(2,2,i+1)
    plt.title(t[i])
    plt.imshow(im[i],cmap="gray")
    plt.axis('off')
plt.show()
```
